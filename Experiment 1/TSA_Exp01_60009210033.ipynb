Detrending using Pandas

```python
from pandas import datetime
from matplotlib import pyplot
def parser(x):
  return datetime.strptime('190'+x, '%Y-%m')
series= pd.read_csv(r'shampoo.csv', index_col=0,parse_dates=True)
X= series.values
diff= list()
for i in range(1, len(X)):
  value= X[i] -X[i-1]
  diff.append(value)
pyplot.figure(figsize=(15,6))
pyplot.plot(diff)
pyplot.title('Detrending using Pandas Differencing', fontsize=16)
pyplot.xlabel('Year')
pyplot.ylabel('Sales exchange rate')
pyplot.show()
```

Importing a Python library called pandas.
Importing a part of the matplotlib library called 'pyplot' for creating plots and graphs.
Defining a function called parser that takes a single argument x. This function will be used to parse dates in a specific format.
It takes the input x, which is expected to be in the format 'YY-MM', and converts it into a datetime object. The format '%Y-%m' specifies that the year should be in a 4-digit format where YY denotes year and MM denotes the month.
The 'series' variable reads the excel file.
The variable 'X' stores the values from the 'series' and holds the exchange rate data.
An empty list called 'diff' where we'll store the differences between consecutive exchange rate values.
Initiates a for loop that iterates over the range of integers from 1 to the length of the 'X' array.
value = X[i] - X[i - 1] will calculates the difference between the exchange rate value at the current index 'i' and the sales at the previous index 'i - 1'. This calculates the first-order difference in the time series data.
diff.append(value) appends the calculated difference 'value' to the 'diff' list.
Ploting the 'diff' list, which contains the differences between consecutive sales exchange rate over time.
Assigning the title of the plot to 'Detrending using Pandas Differencing'.
Seting the label for the x-axis as 'Year'and the y-axis as 'sales exchange rate'.


>The plotted graph provides an initial visual representation of the changes in the sales exchange rate after detrending. It highlights periods of significant increase or decrease in the exchange rate.The detrending process is performed by taking the differences between consecutive data points. It is used in time series analysis to remove the trend component from the data which makes it easier to identify patterns and seasonality.The X-axis represents the year or time period associated with the differences in the sales exchange rate.The Y-axis represents the differences in the sales exchange rate. This indicates how much the sales exchange rate has changed from one period to the next. Positive values indicate an increase in the exchange rate, while negative values indicate a decrease.The primary purpose of differencing is to remove the trend component from the time series data. In this case, positive spikes on the graph represent periods where the sales exchange rate increased significantly compared to the previous period, and negative spikes represent periods where it decreased significantly.

Detrending using Scipy Signal

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy import signal
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
df = pd.read_excel(r'India_Exchange_Rate_Dataset.xls',
index_col=0,parse_dates=True)
detrended = signal.detrend(df.EXINUS.values)
plt.figure(figsize=(15,6))
plt.plot(detrended)
plt.xlabel('EXINUS')
plt.ylabel('Frequency')
plt.title('Detrending using Scipy Signal', fontsize=16)
plt.show()
```

Detrending using HP Filter

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.filters.hp_filter import hpfilter
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
df = pd.read_excel(r'India_Exchange_Rate_Dataset.xls',index_col=0,parse_dates=True)
EXINUS_cycle,EXINUS_trend = hpfilter(df['EXINUS'], lamb=1600)
df['trend'] = EXINUS_trend
detrended = df.EXINUS - df['trend']
plt.figure(figsize=(15,6))
plt.plot(detrended)
plt.title('Detrending using HP Filter', fontsize=16)
plt.xlabel('Year')
plt.ylabel('EXINUS exchange rate')
plt.show()
```
