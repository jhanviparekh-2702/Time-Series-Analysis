Create the three decomposition models from scratch.

```python
import numpy as np
from statsmodels.tsa.seasonal import seasonal_decompose as sm
#Simulating time series components

# setting the range of time series
T_Series = pd.read_csv(r'FB.csv', index_col=0,parse_dates=True)

#create the trend component
Trend = T_Series * 2.75

#plot the trend
plt.plot(T_Series.High, Trend, 'c.')
plt.title("Trend against Time")
```

The output graph shows how the simulated trend component relates to the original 'High' values from the 'T_Series.' Each point on the graph corresponds to a data point in the 'T_Series' on the x-axis and the corresponding trend value on the y-axis.f the points on the graph form a clear, upward-sloping pattern, it indicates a positive trend where the 'Trend' component is consistently higher than the original 'High' values.If the points form a downward-sloping pattern, it indicates a negative trend.If the points are randomly scattered, there may be no discernible trend.The trend is created by scaling the original data by a factor of 2.75, we would expect the 'Trend' component to be consistently higher than the original 'High' values, resulting in an upward-sloping pattern on the graph.

```python
# creating the seasonality component
seasonality = 10 + np.sin(T_Series) * 10 #The sin function as it is the function that can yield those seasonal fluctuations of time series in the best way.

#plot the seasonality trend
plt.plot(T_Series.High,  seasonality, 'b-.')
plt.title("Seasonality against Time")
```

The output graph shows how the simulated seasonality component relates to the original 'High' values from the 'T_Series.' Each point on the graph corresponds to a data point in the 'T_Series' on the x-axis and the corresponding seasonality value on the y-axis.The seasonality component is generated using the sine function, which creates periodic fluctuations. These fluctuations are typically characterized by a regular pattern that repeats over time.The baseline value of 10 added to the sine function's output sets the minimum value of the seasonality component, and scaling it by 10 controls the amplitude (or height) of the fluctuations.If the sine function results in values between -1 and 1, then the seasonality component will fluctuate between 0 and 20.The dashed blue line in the graph represents the seasonality component, and you would expect to see regular oscillations over time.The output graph illustrates how the simulated seasonality component behaves over time when added to the original 'High' values.

```python
#creating the residual component
np.random.seed(10)  # for result reproducibility
residual = np.random.normal(loc=0.0, scale=1, size=len(T_Series)) #Draw random samples from a normal (Gaussian) distribution. loc is the mean, scale is standard deviation

#plot the residual component
plt.plot(T_Series.High, residual, 'r-.')
plt.title("Residuals against Time")
```

The output graph shows how the simulated residual component relates to the original 'High' values from the 'T_Series.' Each point on the graph corresponds to a data point in the 'T_Series' on the x-axis and the corresponding residual value on the y-axis.The residual component is generated as random noise drawn from a normal distribution with a mean of 0 and a standard deviation of 1. This means that the residuals are centered around 0 and have a spread defined by the standard deviation.The red dash-dot line in the graph represents the residual component. Since it's generated randomly, there is no discernible pattern or trend in the residual component.The output graph illustrates the random and unstructured nature of the simulated residual component.

```python
#create an additive model of our time series
additive_Tmodel = Trend + seasonality + residual

#Plot the additive model
plt.plot(T_Series.High, additive_Tmodel, 'k.')
plt.title("Additive Time Series")
```
